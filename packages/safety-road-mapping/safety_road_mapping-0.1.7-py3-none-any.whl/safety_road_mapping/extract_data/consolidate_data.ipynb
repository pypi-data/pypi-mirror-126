```python
import pandas as pd
from glob import glob
```

```python
data_list = glob('./data/*.csv')
```

```python
df_list = []
for data in data_list:
    drop_list = ['feridos_leves', 'ilesos',
                 'ignorados', 'feridos', 'veiculos', 'uso_solo', 'id']
    try:
        df_list.append(pd.read_csv(data, sep=';', low_memory=False, encoding='utf8').drop(columns=drop_list))
    except:
        df_list.append(pd.read_csv(data, sep=';', low_memory=False, encoding='latin').drop(columns=drop_list))
```

```python
len(df_list)
```

```python
all_data = pd.concat(df_list)
all_data = all_data.drop(columns=['regional', 'delegacia', 'uop'])
```

```python
all_data.head()
```

```python
all_data = all_data.dropna(subset=['latitude', 'longitude'])
```

```python
len(all_data)
```

```python
all_data.loc[:, 'data_inversa'] = all_data['data_inversa'].apply(lambda x: '-'.join(x.split('/')[::-1]))
all_data.loc[:, 'data_inversa'] = all_data['data_inversa'].str.replace('16-', '2016-')
all_data.loc[:, 'data_inversa'] = pd.to_datetime(all_data['data_inversa'], format='%Y-%m-%d')
all_data = all_data.sort_values('data_inversa')
all_data = all_data.drop(columns='ano')
```

```python
all_data.to_csv('accidents_data.csv', index=False, encoding='latin1', sep=';')
```
