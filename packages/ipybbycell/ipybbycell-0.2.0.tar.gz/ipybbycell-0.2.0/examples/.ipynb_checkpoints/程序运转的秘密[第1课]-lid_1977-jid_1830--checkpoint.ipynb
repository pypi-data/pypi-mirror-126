```python
//platform-lock 
srand(time(0));
int answer = rand() % 10 + 1;       
int user = 0;
int cnt = 0;
while (1) {
    cout << "请输入你猜的数字（1到10之间）：" << endl;     
    cin >> user;
    cnt++;
    if (user > answer) 
        cout << user << "大了" << endl;
    if (user < answer)
        cout << user << "小了" << endl;                
    if (user == answer) {
        cout << "猜对了" << endl;
        cout << "你一共猜了" << cnt << "次。";
        break;
    }
}
```
